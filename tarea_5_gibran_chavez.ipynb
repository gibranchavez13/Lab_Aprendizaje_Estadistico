### Importación de librerías

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

### Carga de archivo

```python
df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
df.head()
```

## Parte 1

Realizar una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considerar todos los demás factores como numéricos/ordinales.

```python
X1_1 = df.drop(['mpg', 'model'], axis=1)
y1_1 = df['mpg']
```

Calcular el R2 e interpretar los signos de los betas.

```python
X1_1_w_constant = sm.add_constant(X1_1)
model1_1 = sm.OLS(y1_1, X1_1_w_constant)
results1_1 = model1_1.fit()
print(results1_1.summary())
```

Sin tomar en cuenta los p-values, algunas $\beta_S$ son positivas y negativas, esto quiere decir que por cada unidad dedicada a cada factor, el resultado de mpg es afectado, en este caso, la variable con más peso positivamente es 'am' y negativamente es 'wt'.

Realizar un train-test-split donde se use el 40% de los datos para entrenar. Calcular el R2 de entrenamiento y de prueba.

```python
X_train1_1, X_test1_1, y_train1_1, y_test1_1 = train_test_split(X1_1, y1_1, test_size=0.4, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train1_1, y_train1_1)
y_pred1_1 = lr.predict(X_test1_1)
r2_1_1train = lr.score(X_train1_1, y_train1_1)
r2_1_1test = lr.score(X_test1_1, y_test1_1)
print(f'R^2 train: {r2_1_1train}')
print(f'R^2 test: {r2_1_1test}')
```

Añadir regularización L2 con un hiperparámetro lambda. Cambiar este valor y comparar con varios distintos los R2 de entrenamiento y de prueba.

```python
scaler = StandardScaler()

X_train1_1_scaled = scaler.fit_transform(X_train1_1)
X_test1_1_scaled = scaler.transform(X_test1_1)
```

```python
alphas = [0.01, 0.1, 1, 5, 10, 50, 100]
for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train1_1_scaled, y_train1_1)
    r2_ridge_train = ridge.score(X_train1_1_scaled, y_train1_1)
    r2_ridge_test = ridge.score(X_test1_1_scaled, y_test1_1)
    print(f'Alpha: {alpha} - R^2 train: {r2_ridge_train} - R^2 test: {r2_ridge_test}')
```

A medida que ponemos un $\alpha$ más grande, la $R^2$ desciende en el train pero asciende en el test.

## Parte 2

Realizar una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Crear columnas dummies para los factores 'cyl', 'gear' y 'carb'.

```python
X2 = df.drop(['mpg', 'model'], axis=1)
y2 = df['mpg']
```

```python
X2_1_dummies = pd.get_dummies(X2, columns=['cyl', 'gear', 'carb'], dtype=int)
X2_1_dummies.head()
```

Calcular el R2 e interpretar los signos de los betas.

```python
X2_dummies_w_constant = sm.add_constant(X2_1_dummies)
model2_1 = sm.OLS(y2, X2_dummies_w_constant)
results2_1 = model2_1.fit()
print(results2_1.summary())
```

Sin tomar en cuenta los p-values, vemos que con dummies, las $\beta_S$ cambian cuando cambiamos las variables con factores numéricos/ordinales por columnas con booleanos, esto indica que con esto el modelo se puede calcular de una forma diferente una y. Vemos que la $\beta$ con más peso positivo es 'carb_8' y negativo es 'wt'. Sin tomar en cuenta la constante.

Además se ve que la $R^2$ se elevó un poco, haciéndolo más preciso.

Realizar un train-test-split donde se use el 40% de los datos para entrenar. Calcular el R2 de entrenamiento y de prueba.

```python
X_train2_1, X_test2_1, y_train2_1, y_test2_1 = train_test_split(X2_1_dummies, y2, test_size=0.4, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train2_1, y_train2_1)
y_pred2_1 = lr.predict(X_test2_1)
r2_2_1train = lr.score(X_train2_1, y_train2_1)
r2_2_1test = lr.score(X_test2_1, y_test2_1)
print(f'R^2 train: {r2_2_1train}')
print(f'R^2 test: {r2_2_1test}')
```

```python
X2_2dummies_w_constant = sm.add_constant(X2_2_dummies)
model2_2 = sm.OLS(y2_2, X2_2dummies_w_constant)
results2_2 = model2_2.fit()
print(results2_2.summary())
```

Esto indica que con esto el modelo puede ser más exacto a la hora de calcular una y. Vemos que la $\beta$ con más peso positivo es 'gear_4' y negativo es 'am'. Sin tomar en cuenta la constante.


```python
X_train2_2, X_test2_2, y_train2_2, y_test2_2 = train_test_split(X2_2_dummies, y2_2, test_size=0.4, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train2_2, y_train2_2)
y_pred2_2 = lr.predict(X_test2_2)
r2_2_2train = lr.score(X_train2_2, y_train2_2)
r2_2_2test = lr.score(X_test2_2, y_test2_2)
print(f'R^2 train: {r2_2_2train}')
print(f'R^2 test: {r2_2_2test}')
```

Comparar los R2 de los ejercicios 1.1 & 2.1.

```python
print(f'R^2 all without dummies: {results1_1.rsquared}')
print(f'R^2 train without dummies: {r2_1_1train}')
print(f'R^2 test without dummies: {r2_1_1test}')
print(f'R^2 all with dummies: {results2_1.rsquared}')
print(f'R^2 train with dummies: {r2_2_1train}')
print(f'R^2 test with dummies: {r2_2_1test}')
```

Comparar los R2 de los ejercicios 1.2 & 2.2.

```python
print(f'R^2 all without dummies: {results1_2.rsquared}')
print(f'R^2 train without dummies: {r2_1_2train}')
print(f'R^2 test without dummies: {r2_1_2test}')
print(f'R^2 all with dummies: {results2_2.rsquared}')
print(f'R^2 train with dummies: {r2_2_2train}')
print(f'R^2 test with dummies: {r2_2_2test}')
```

En ambas comparaciones vemos claramente como la $R^2$ se incrementó al manejar dummies en algunas variables con factores numéricos/ordinales, lo que indica que siendo más excato a la hora de recabar información, se puede incrementar la precisión de un modelo.

La unica $R^2$ que bajó al manejar dummies, es en los modelos $Ridge$, esto porque como este modelo penaliza más, al aumentar en número de predictores con los dummies, el efecto de una $\beta$ se reparte en varias al separarlas pro dummies, y $Ridge$ penaliza todas ellas.
