### 1.

Usando propiedades estadísticas básicas de la varianza, así como cálculo de una sola variable, derive la ecuación (5.6). En otras palabras, demuestre que el valor de $\alpha$ dado por (5.6) efectivamente minimiza:

$$
\text{Var}(\alpha X + (1 - \alpha)Y)
$$


Partimos de:

$$
\text{Var}(\alpha X + (1-\alpha)Y)
$$

Usamos la propiedad:

$$
\text{Var}(aX+bY)=a^2\text{Var}(X)+b^2\text{Var}(Y)+2ab\text{Cov}(X,Y)
$$

Entonces:

$$
\text{Var}(\alpha X + (1-\alpha)Y)
=
\alpha^2\sigma_X^2+(1-\alpha)^2\sigma_Y^2+2\alpha(1-\alpha)\sigma_{XY}
$$

---

### Expansión

Sabemos que:

$$
(1-\alpha)^2 = 1 - 2\alpha + \alpha^2
$$

y:

$$
2\alpha(1-\alpha)\sigma_{XY}
=
2\alpha\sigma_{XY}-2\alpha^2\sigma_{XY}
$$

Sustituyendo:

$$
V(\alpha)
=
\alpha^2\sigma_X^2
+
\sigma_Y^2
-
2\alpha\sigma_Y^2
+
\alpha^2\sigma_Y^2
+
2\alpha\sigma_{XY}
-
2\alpha^2\sigma_{XY}
$$

---

### Agrupación de términos

$$
V(\alpha)
=
\alpha^2(\sigma_X^2+\sigma_Y^2-2\sigma_{XY})
+
\alpha(-2\sigma_Y^2+2\sigma_{XY})
+
\sigma_Y^2
$$

---

### Derivada

Derivamos respecto a $\alpha$:

$$
\frac{dV}{d\alpha}
=
2\alpha(\sigma_X^2+\sigma_Y^2-2\sigma_{XY})
+
(-2\sigma_Y^2+2\sigma_{XY})
$$

Igualamos a cero:

$$
2\alpha(\sigma_X^2+\sigma_Y^2-2\sigma_{XY})
-
2\sigma_Y^2
+
2\sigma_{XY}
=
0
$$

Dividimos entre 2:

$$
\alpha(\sigma_X^2+\sigma_Y^2-2\sigma_{XY})
-
\sigma_Y^2
+
\sigma_{XY}
=
0
$$

---

### Despeje de $\alpha$

$$
\alpha(\sigma_X^2+\sigma_Y^2-2\sigma_{XY})
=
\sigma_Y^2-\sigma_{XY}
$$

$$
\alpha =
\frac{\sigma_Y^2 - \sigma_{XY}}
{\sigma_X^2+\sigma_Y^2-2\sigma_{XY}}
$$

Además, para verificar que el valor obtenido corresponde a un mínimo, analizamos la segunda derivada de la varianza:

$$
\frac{d^2V}{d\alpha^2}
=
2(\sigma_X^2+\sigma_Y^2-2\sigma_{XY})
$$

Observamos que:

$$
\sigma_X^2+\sigma_Y^2-2\sigma_{XY}
=
Var(X-Y)
$$

Como una varianza siempre es mayor o igual que cero, entonces:

$$
\frac{d^2V}{d\alpha^2} \geq 0
$$

---


### Conclusión

El valor obtenido de $\alpha$ es:

$$
\alpha =
\frac{\sigma_Y^2-\sigma_{XY}}
{\sigma_X^2+\sigma_Y^2-2\sigma_{XY}}
$$

Este valor minimiza la varianza de la combinación lineal:

$$
\alpha X + (1-\alpha)Y
$$

porque se obtuvo al derivar la función de varianza e igualar a cero. Además, la segunda derivada es:

$$
\frac{d^2V}{d\alpha^2}
=
2(\sigma_X^2+\sigma_Y^2-2\sigma_{XY})
=
2Var(X-Y)
$$

Como una varianza siempre es mayor o igual que cero, la función es convexa respecto a $\alpha$. Por lo tanto, el punto crítico encontrado corresponde a un mínimo de la varianza.

#### (a)

¿Cuál es la probabilidad de que la primera observación bootstrap **no sea** la j-ésima observación de la muestra original? Justifique su respuesta.

La probabilidad de que la primera observación bootstrap no sea la j-ésima observación original es:

$$
1 - \frac{1}{n}
$$

Esto se debe a que la probabilidad de elegir la j-ésima observación es:

$$
\frac{1}{n}
$$

Por lo tanto, la probabilidad de no elegirla es:

$$
1 - \frac{1}{n}
$$

---

#### (b)

¿Cuál es la probabilidad de que la segunda observación bootstrap **no sea** la j-ésima observación de la muestra original?

Como cada selección se hace con reemplazo, la probabilidad no cambia de una extracción a otra.

Por lo tanto, la probabilidad de que la segunda observación bootstrap no sea la j-ésima observación original también es:

$$
1 - \frac{1}{n}
$$


---

#### (c)

Argumente que la probabilidad de que la j-ésima observación **no esté** en la muestra bootstrap es:

$$
(1 - \frac{1}{n})^n
$$

Para que la j-ésima observación no aparezca en toda la muestra bootstrap, debe no salir en ninguna de las $n$ selecciones.

Como las selecciones son independientes, multiplicamos las probabilidades:

$$
P(\text{j no aparece}) =
\left(1-\frac{1}{n}\right)^n
$$

Por lo tanto, la probabilidad de que la j-ésima observación sí aparezca al menos una vez es:

$$
P(\text{j aparece}) =
1-\left(1-\frac{1}{n}\right)^n
$$

### 3.

Ahora revisamos la validación cruzada k-fold.

#### (a)

Explique cómo se implementa la validación cruzada k-fold.

La validación cruzada k-fold consiste en dividir el conjunto de datos en $k$ partes aproximadamente del mismo tamaño, llamadas **folds**.

El procedimiento es el siguiente:

1. Se divide el conjunto de datos en $k$ grupos.
2. Se toma un fold como conjunto de validación.
3. Los otros $k-1$ folds se usan como conjunto de entrenamiento.
4. Se ajusta el modelo usando los $k-1$ folds de entrenamiento.
5. Se calcula el error del modelo sobre el fold de validación.
6. Este proceso se repite $k$ veces, usando cada fold una vez como validación.
7. Finalmente, se promedian los $k$ errores obtenidos.

La estimación del error de prueba se puede expresar como:

$$
CV_{(k)} = \frac{1}{k}\sum_{i=1}^{k} MSE_i
$$

donde $MSE_i$ representa el error obtenido en el fold $i$.

En clasificación, en lugar del $MSE$, se puede usar una métrica como accuracy, recall, precision, F1-score o ROC-AUC, dependiendo del objetivo del modelo.


----

#### (b)

¿Cuáles son las ventajas y desventajas de k-fold en comparación con:

1. El método de conjunto de validación.

2. LOOCV (*Leave-One-Out Cross-Validation*).

#### i. Comparación con el método de conjunto de validación

El método de conjunto de validación divide los datos una sola vez en dos partes:

- Conjunto de entrenamiento.
- Conjunto de validación.

Por ejemplo, se puede usar 70% de los datos para entrenamiento y 30% para validación.

La principal desventaja de este método es que el resultado puede depender mucho de cómo se haya hecho la división inicial. Si por azar el conjunto de validación contiene observaciones poco representativas, la estimación del error puede ser poco confiable.

En cambio, k-fold usa todos los datos tanto para entrenamiento como para validación, aunque en diferentes iteraciones. Cada observación se usa una vez para validar y varias veces para entrenar.

##### Ventajas de k-fold frente al conjunto de validación

- Usa mejor los datos disponibles.
- Es menos sensible a una sola división aleatoria.
- Suele dar una estimación más estable del error de prueba.
- Todas las observaciones participan en entrenamiento y validación.

##### Desventajas de k-fold frente al conjunto de validación

- Es más costoso computacionalmente.
- El modelo debe entrenarse $k$ veces.
- Puede tardar más si el modelo es complejo o si el dataset es grande.

---

#### ii. Comparación con LOOCV

LOOCV, o *Leave-One-Out Cross-Validation*, es un caso especial de validación cruzada donde:

$$
k = n
$$

Es decir, si el conjunto de datos tiene $n$ observaciones, el modelo se entrena $n$ veces. En cada iteración se deja fuera una sola observación como validación y se entrena con las $n-1$ observaciones restantes.

##### Ventajas de k-fold frente a LOOCV

- Es mucho más rápido computacionalmente.
- Solo requiere entrenar el modelo $k$ veces, no $n$ veces.
- Es más práctico para conjuntos de datos grandes.
- Puede tener menor varianza en la estimación del error de prueba.

##### Desventajas de k-fold frente a LOOCV

- En cada iteración usa menos datos para entrenar que LOOCV.
- Puede tener un poco más de sesgo que LOOCV.
- El resultado puede depender de cómo se formen los folds.

---

### Conclusión

La validación cruzada k-fold es un punto medio entre el método de conjunto de validación y LOOCV.

Comparada con el método de conjunto de validación, suele ser más estable porque no depende de una sola partición aleatoria de los datos.

Comparada con LOOCV, suele ser mucho más eficiente computacionalmente y más práctica para problemas reales.

Por esta razón, valores como:

$$
k = 5
$$

o

$$
k = 10
$$

son muy comunes en machine learning.

En conclusión, \(k\)-fold cross-validation representa un punto medio entre el método de validación simple y LOOCV. A diferencia del validation set approach, utiliza todas las observaciones tanto para entrenamiento como para validación en algún momento del proceso, por lo que suele producir estimaciones más estables del error de prueba.

Además, aunque LOOCV entrena cada modelo con \(n-1\) observaciones y por ello suele tener bajo sesgo, también puede presentar mayor varianza, ya que los conjuntos de entrenamiento son muy parecidos entre sí. En cambio, \(k\)-fold cross-validation, especialmente con valores como \(k=5\) o \(k=10\), suele ofrecer un mejor equilibrio entre sesgo, varianza y costo computacional.

### 4.

Suponga que usamos algún método de aprendizaje estadístico para hacer una predicción de la respuesta $Y$ para un valor particular del predictor $X$. Describa cuidadosamente cómo podríamos estimar la desviación estándar de nuestra predicción.

---

Cuando hacemos una predicción $\hat{Y}$, no solo queremos el valor estimado, sino también la incertidumbre asociada a esa predicción.

Esta incertidumbre se mide mediante la **desviación estándar de la predicción**, también conocida como:

- Error estándar de predicción  
- Incertidumbre del modelo  

---

Para estimar la desviación estándar de una predicción en un punto específico \($x_0$\), puede utilizarse el método bootstrap.

El procedimiento sería el siguiente:

1. Generar \(B\) muestras bootstrap a partir del conjunto de datos original, tomando observaciones con reemplazo.
2. Ajustar el modelo estadístico en cada muestra bootstrap.
3. Para cada modelo ajustado, calcular la predicción en el mismo punto \($x_0$\).
4. Obtener una colección de predicciones bootstrap:

$$
\hat{y}_0^{*(1)}, \hat{y}_0^{*(2)}, \ldots, \hat{y}_0^{*(B)}
$$

5. Calcular la desviación estándar de esas predicciones:

$$
SE(\hat{y}_0)
=
\sqrt{
\frac{1}{B-1}
\sum_{b=1}^{B}
\left(
\hat{y}_0^{*(b)}-\bar{\hat{y}}_0^*
\right)^2
}
$$

donde:

$$
\bar{\hat{y}}_0^*
=
\frac{1}{B}
\sum_{b=1}^{B}
\hat{y}_0^{*(b)}
$$

Este método es útil porque no depende de una fórmula cerrada específica del modelo. Por ello, puede aplicarse no solo a regresión lineal, sino también a modelos no lineales o métodos de aprendizaje estadístico más complejos.

La validación cruzada sirve principalmente para estimar el error de predicción promedio del modelo, pero no es el método más directo para estimar la desviación estándar de una predicción puntual. Por esa razón, bootstrap es más apropiado para este caso.

````python
errors = np.zeros(len(X))
````

````python
for i in range(len(X)):

    X_train = X.drop(index=X.index[i])
    y_train = y.drop(index=y.index[i])

    X_test = X.iloc[[i]]
    y_test = y.iloc[i]

    log_r = LogisticRegression()
    log_r.fit(X_train, y_train)

    proba_up = log_r.predict_proba(X_test)[0][1]

    if proba_up > 0.5:
        prediccion = 1
    else:
        prediccion = 0

    if prediccion != y_test:
        errors[i] = 1
    else:
        errors[i] = 0
````

````python
from sklearn.preprocessing import PolynomialFeatures # PolynomialFeatures ayuda a crear el polinomio sin la necesidad de hacerlo a mano
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import LeaveOneOut, cross_val_score # LeaveOneOut ayuda a hacer el LOOCV sin necesidad de crear un ciclo para quedarnos con n-1 muestras
````

````python
X = df[['x']]
y = df['y']
````

````python
loo = LeaveOneOut()

loo_errors = []

grados = [1, 2, 3, 4]
````

````python
for grado in grados:
    pf = PolynomialFeatures(degree=grado, include_bias=False)
    lr = LinearRegression()

    pip = Pipeline(
        steps=[
            ('grado_polinomio', pf),
            ('model', lr)
        ]
    )

    avg_score = cross_val_score(
        pip, 
        X,
        y,
        cv=loo,
        scoring='neg_mean_squared_error'
    ).mean()
    
    loo_errors.append((-avg_score))
````
