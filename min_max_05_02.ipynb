## Estandarizar con standard (z)

```python
# Antes
for x in X.columns:
    plt.hist(X[x], bins=5)
    plt.title(x + ' antes de la normalización')
    plt.show() 
```

```python
X_scaled2 = X - X.mean()
X_scaled2 = X_scaled2 / X_scaled2.std()

X_scaled2.describe()
```

```python
for x in X_scaled2:    
    plt.hist(X_scaled2[x], bins=5)
    plt.title(x)
    plt.show()
```
