# Actividad 1 - $Bootstrap$

1. Tomar dataset $\text{Motor Trend Road}$

2. Realizar regresión lineal sobre $\hat{mpg}$ = $\hat{\beta_0}$ + $\hat{\beta_1} (hp)$ + $\hat{\beta_2} (qsec)$


    2.1 Calcular intervalos de confianza de los $\hat{\beta_s}$

3. Crear 1000 muestras de bootstrap con $m$ observaciones con remplazo

    3.1 Realizar regresión lineal sobre $\hat{mpg}$ = $\hat{\beta_0}$ + $\hat{\beta_1} (hp)$ + $\hat{\beta_2} (qsec)$


    3.2 Calcular $\bar{\beta_x}$ y $\sigma_b$, abrir intervalos de confianza

4. Comparar resultados entre (2) y (3)

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
df = pd.read_excel('Motor Trend Car Road Tests.xlsx')
df.head()
```

```python
x_cols = ['hp', 'qsec']
target = 'mpg'
```

```python
X = df[x_cols]
X.head()
```

```python
y = df[target]
y
```

### Realizar regresión lineal sobre $\hat{mpg}$ = $\hat{\beta_0}$ + $\hat{\beta_1} (hp)$ + $\hat{\beta_2} (qsec)$ e intervalos de confianza


```python
X_w_const = sm.add_constant(X)
```

```python
model = sm.OLS(y, X_w_const)
results = model.fit()
print(results.summary())
```

### Crear 1000 muestras de bootstrap con $m$ observaciones con remplazo

```python
muestras_x = []
muestras_y = []

for i in range(1000):
    bootstrap_x = X.sample(n=len(X), replace=True, random_state=i)
    bootstrap_y = y.sample(n=len(y), replace=True, random_state=i)

    muestras_x.append(bootstrap_x)
    muestras_y.append(bootstrap_y)
```

### Realizar regresión lineal sobre $\hat{mpg}$ = $\hat{\beta_0}$ + $\hat{\beta_1} (hp)$ + $\hat{\beta_2} (qsec)$ con cada muestra


```python
betas_0 = []
betas_1 = []
betas_2 = []


for i in range(1000):
    x_w_const_i = sm.add_constant(muestras_x[i])

    model_i = sm.OLS(muestras_y[i], x_w_const_i)
    results_i = model_i.fit()

    betas_0.append(results_i.params[0])
    betas_1.append(results_i.params[1])
    betas_2.append(results_i.params[2])
```

### Calcular $\bar{\beta_x}$ y $\sigma_b$

### Abrir intervalos de confianza


```python
betas_0_avg = np.mean(betas_0)
betas_0_desv = np.std(betas_0)

betas_1_avg = np.mean(betas_1)
betas_1_desv = np.std(betas_1)

betas_2_avg = np.mean(betas_2)
betas_2_desv = np.std(betas_2)
```

Para $\beta_0$

```python
b0_lower = betas_0_avg - 2*betas_0_desv
b0_upper = betas_0_avg + 2*betas_0_desv

print(f'Intervalo para b_0: [{b0_lower}, {b0_upper}]')
```

Para $\beta_1$

```python
b1_lower = betas_1_avg - 2*betas_1_desv
b1_upper = betas_1_avg + 2*betas_1_desv

print(f'Intervalo para b_0: [{b1_lower}, {b1_upper}]')
```

Para $\beta_2$

```python
b2_lower = betas_2_avg - 2*betas_2_desv
b2_upper = betas_2_avg + 2*betas_2_desv

print(f'Intervalo para b_0: [{b2_lower}, {b2_upper}]')
```

### Comparar resultados entre (2) y (3)

Promedios

```python
print(f'Intervalo para b_0: [{b0_lower}, {b0_upper}]')
print(f'Intervalo para b_0: [{b1_lower}, {b1_upper}]')
print(f'Intervalo para b_0: [{b2_lower}, {b2_upper}]')
```

# Actividad 2 - $Aggregating$

1. A partir de la matriz de diseño $X$ con todos los datos, crear 1000 modelos con 3 columnas al azar

2. Realizar train-test 50/50

3. Entrenar los 1000 modelos con train

4. Predecir con test ($\hat{y}_{test}$)

5. Calcular ${R^2}_{test}$

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df.head()
```

```python
matriz_diseno = df.drop(['model', 'mpg'], axis=1)
matriz_diseno.head()
```

```python
y = df[target]
y.head()
```

```python
len(matriz_diseno.columns)
```

### Train-Test 50/50

```python
X_train, X_test, y_train, y_test = train_test_split(matriz_diseno, y, test_size=0.5, random_state=42)
```

### Creación de 1000 modelos con 3 columnas al azar

```python
vars = []
models = []
```

```python
for _ in range(1000):
    var = np.random.choice(matriz_diseno.columns, 3, replace=False)
    
    vars.append(var)

    X_w_random_vars = X_train[var]

    lr = LinearRegression()

    model = lr.fit(X_w_random_vars, y_train)

    models.append(model)   
```

### Predicciones con test ($\hat{y}_{test}$)

```python
y_preds = []
```

```python
for i in range(1000):
    X_test_i = X_test[vars[i]]

    y_pred_i = models[i].predict(X_test_i)

    y_preds.append(y_pred_i)
```

### $\bar{\hat{y}}$

```python
y_pred_avg = np.mean(y_preds, axis=0)
y_pred_avg
```

### $R^2$

```python
r2_score(y_test, y_pred_avg)
```
