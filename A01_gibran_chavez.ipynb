### Realiza regresiones polinomiales de orden 1, 3, 5 y 7.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data_tablets = pd.read_csv('bateria_tableta.txt', sep=',')
data_tablets.head()
```

### Para bateria_tabletas.txt

Regresión orden 1

```python
y_tablets = np.array(data_tablets['duracion'])

X1_tablets = np.array([data_tablets['carga']]).T
```

```python
lr1_tablets = LinearRegression() 
lr1_tablets.fit(X1_tablets, y_tablets)
lr1_tablets.intercept_
```

```python
lr1_tablets.coef_
```

```python
lr1_tablets.score(X1_tablets, y_tablets)
```

Regresión orden 3

```python

X3_tablets = np.array([data_tablets['carga']**i for i in range(1, 4)]).T
```

## Comparación de cada R^2


```python
print('R^2 score para cada regresion del dataset de Tablets:')
print(f"R^2 score para regresion de grado 1: {lr1_tablets.score(X1_tablets, y_tablets)}")
print(f"R^2 score para regresion de grado 3: {lr3_tablets.score(X3_tablets, y_tablets)}")
print(f"R^2 score para regresion de grado 5: {lr5_tablets.score(X5_tablets, y_tablets)}")
print(f"R^2 score para regresion de grado 7: {lr7_tablets.score(X7_tablets, y_tablets)}, \n")

print('R^2 score para cada regresion del dataset de Salarios:')
print(f"R^2 score para regresion de grado 1: {lr1_salarios.score(X1_salarios, y_salarios)}")
print(f"R^2 score para regresion de grado 3: {lr3_salarios.score(X3_salarios, y_salarios)}")
print(f"R^2 score para regresion de grado 5: {lr5_salarios.score(X5_salarios, y_salarios)}")
print(f"R^2 score para regresion de grado 7: {lr7_salarios.score(X7_salarios, y_salarios)}")
```

Mientras más sea el grado de la regresión, la R^2 se vuelve más grande considerablemente, obteniedo mayor precisión.

## Crea una gráfica con una nube de dispersión (datos originales) y tus modelos

Para el dataset de Tablets

```python
plt.figure(figsize=(20, 10)); plt.grid()
plt.scatter(data_tablets['carga'], data_tablets['duracion'], color='black', label='Datos reales')
plt.scatter(data_tablets['carga'], lr1_tablets.predict(X1_tablets), color='orange', label='Regresion grado 1')
plt.scatter(data_tablets['carga'], lr3_tablets.predict(X3_tablets), color='red', label='Regresion grado 3')
plt.scatter(data_tablets['carga'], lr5_tablets.predict(X5_tablets), color='green', label='Regresion grado 5')
plt.scatter(data_tablets['carga'], lr7_tablets.predict(X7_tablets), color='blue', label='Regresion grado 7')
plt.show()
```
