# **Marco teórico**

##*Regresión logística*
La regresión logística es un modelo estadístico de clasificación supervisada que estima la probabilidad de que una observación pertenezca a una clase determinada, utilizando la función sigmoide para mapear cualquier valor real al intervalo (0, 1). La función sigmoide tiene forma de "S" y transforma cualquier número (sin importar qué tan grande o pequeño sea) en un valor entre 0 y 1 y poder interpretarlo como una probabilidad.

Ahora bien, para evitar el sobreajuste se incorporan términos de regularización como:L1 (Lasso) y L2 (Ridge); los cuales  penalizan la magnitud de los coeficientes y mejoran la capacidad de generalización del modelo ante datos no vistos. La regularización L1 favorece soluciones dispersas al llevar algunos coeficientes exactamente a cero, mientras que L2 distribuye la penalización de forma más uniforme entre todos los parámetros y los coeficientes sólo tienden a cero.

Se utiliza una nueva forma de transformación de potencia, ya que toma una variable que puede estar sesgada lo que nos llevan a inducir si es que es muy asimétrica o con una varianza inestable, y la transforma para que quede más cercana a una distribución normal o al menos más simétrica. SciPy lo describe justamente como una transformación para mejorar normalidad o simetría, y scikit-learnindica que estas transformaciones ayudan a estabilizar la varianza y minimizar la asimetría.

Logramos incluir un BaseEstimator ya que nos es útil porque es la clase base para todos los estimadores. Al heredar de ella, una clase obtiene implementaciones por defecto para manejar parámetros, representación del objeto y otras utilidades del ecosistema de sklearn. En particular, la documentación nos destaca que provee implementaciones por defecto para establecer y obtener parámetros, que es precisamente lo que necesitan herramientas como GridSearchCV y afines.

Otra herramienta utilizada fue ClassifierMixin, aunque lo manejamos como  scikit-learn que MLPWrapperdebe tratarse como un modelo de clasificación, no como un regresor. Eso es importante porque muchas utilidades internas y muchos scorers cambian su comportamiento dependiendo de si el estimador es clasificador o regresor.

## *Máquinas de vectores de soporte*

Las Máquinas de Vectores de Soporte son algoritmos de clasificación que buscan trazar la mejor línea o superficie posible para separar dos clases en los datos. A esta línea o superficie divisoria se le llama hiperplano o frontera de decisión: en dos dimensiones es una línea recta, en tres dimensiones es un plano y en más dimensiones sigue el mismo principio. El algoritmo elige el hiperplano que deja el mayor espacio posible entre él y los puntos más cercanos de cada clase, maximizando así el margen de separación.

Cuando los datos no pueden separarse con una línea recta, las SVM utilizan un kernel, el cual transforma los datos a un espacio de mayor dimensión donde sí es posible separarlos linealmente, sin necesidad de realizar esa transformación de forma explícita y costosa. Por ejemplo, puntos que forman círculos en 2D pueden volverse separables al proyectarlos a 3D. Es importante mencionar que el  kernel de base radial (RBF) es uno de los más utilizados por su flexibilidad para adaptarse a distintas formas de datos,

##*Redes neuronales (Multi-layer perceptron)*
El Perceptrón Multicapa es una red neuronal compuesta por una capa de entrada, una o más capas ocultas y una capa de salida, donde cada neurona aplica una función de activación no lineal sobre la combinación lineal ponderada de sus entradas. Los MLP son aproximadores universales de funciones, ya que son capaces de capturar relaciones no lineales complejas entre variables de entrada y la variable objetivo.

El aprendizaje ocurre mediante la retropropagación, es decir, la red hace una predicción, calcula qué tan equivocada estuvo (el error) y luego recorre la red de atrás hacia adelante ajustando cada peso en proporción a cuánto contribuyó a ese error. Este proceso se repite miles de veces hasta que el modelo mejora progresivamente sus predicciones. La profundidad de la red, el número de neuronas por capa y la función de activación son decisiones de diseño que afectan directamente la capacidad del modelo.

## *Kernels*
En un aprendizaje automático, un kernel es una función que mide la similitud entre dos observaciones en un espacio transformado de mayor dimensión, sin necesidad de calcular esa transformación de forma explícita. En otras palabras, es una función que permite que algoritmos como las SVM operen como si los datos estuvieran en un espacio más complejo, sin el costo computacional que eso implicaría.

Los kernels más utilizados incluyen el lineal, el polinomial y el de base radial (RBF). El RBF es el más popular por su parámetro gamma, el cual controla el radio de influencia de cada punto de entrenamiento, permitiendo ajustar la suavidad de la frontera de decisión. La elección adecuada del kernel determina la capacidad expresiva del modelo y su comportamiento ante distintas geometrías de los datos.

##*Métricos para clasificación*
La evaluación del desempeño de modelos de clasificación requiere métricas que vayan más allá de accuracy, especialmente cuando las clases están desbalanceadas. La matriz de confusión permite derivar indicadores como la precisión, el recall y el F1-score que combinan los conteos de verdaderos positivos, falsos positivos y falsos negativos para ofrecer una visión más completa del comportamiento del clasificador. Además, el área bajo la curva ROC (AUC-ROC) evalúa la capacidad discriminativa del modelo a lo largo de distintos umbrales de decisión, siendo independiente del umbral elegido y útil para comparar modelos entre sí.

## *Hiperparámetros*
Los hiperparámetros son configuraciones del modelo que el programador debe definir antes de entrenar, ya que el algoritmo no puede aprenderlos por sí solo a partir de los datos como la tasa de aprendizaje (qué tan grandes son los ajustes en cada paso del entrenamiento), el parámetro C en SVM (qué tan estricto es el margen de separación) o el número de capas ocultas en un MLP.

Para encontrar los mejores valores, una estrategia común es el grid search (búsqueda en cuadrícula), que consiste en probar todas las combinaciones posibles de un conjunto predefinido de valores. Por ejemplo, si se quiere ajustar en un rango específico, el grid search entrena y evalúa un modelo para cada una de las combinaciones posibles, seleccionando la que mejor desempeño tenga. Aunque garantiza explorar todas las opciones definidas, puede volverse costoso computacionalmente cuando el espacio de búsqueda es grande.

## *Optimización Bayesiana*
La optimización bayesiana es una estrategia para encontrar los mejores hiperparámetros sin tener que probarlos todos. Construye un modelo probabilístico que predice qué tan bueno será el desempeño del modelo para una combinación de hiperparámetros dada y usa esa predicción para decidir cuál combinación vale la pena evaluar a continuación. Para guiar esa decisión utiliza una función de adquisición, es una regla que balancea dos objetivos: explorar zonas del espacio de búsqueda que aún no se conocen bien y aprovechar las zonas donde ya se encontraron buenos resultados.

Este balance entre exploración y explotación reduce significativamente el número de evaluaciones necesarias para encontrar configuraciones óptimas, lo que lo convierte en el método preferido cuando el entrenamiento del modelo es costoso computacionalmente, como las redes neuronales profundas o SVM sobre conjuntos de datos de gran tamaño.


# **Análisis de dataset**

## *¿De dónde viene?*
Este dataset fue creado por el Dr. William H. Wolberg en la Universidad de Wisconsin-Madison, y hoy en día es uno de los conjuntos de datos más usados en el mundo del aprendizaje automático. Los datos vienen de biopsias reales de pacientes: específicamente de un procedimiento llamado aspiración con aguja fina, donde se extrae una pequeña muestra de tejido de una masa en la mama y se digitaliza en imagen para analizarla. Esas imágenes fueron procesadas para medir características de los núcleos de las células, y ese trabajo se convirtió en lo que hoy conocemos como este archivo.

## *¿Qué contiene?*
El archivo tiene 569 registros, uno por cada paciente. Para cada uno se registraron 30 variables numéricas que describen la forma, tamaño y textura de los núcleos celulares observados en la biopsia. Esas 30 variables no son 30 cosas distintas: en realidad son 10 características medidas de tres formas diferentes, el promedio, el error estándar y el valor más extremo encontrado. Además, cada registro tiene un diagnóstico: M si el tumor resultó maligno, y B si resultó benigno.

## *¿Qué información dan las muestras?*
Cada fila del archivo nos está contando la historia morfológica de un tumor. Variables como el radio, el perímetro o el área nos dicen qué tan grande es el núcleo celular. La textura nos habla de qué tan irregular se ve su superficie. La concavidad y los puntos cóncavos describen qué tan "arrugado" o hundido está el contorno.
La simetría dice qué tan parecidas son las dos mitades del núcleo, y la dimensión fractal mide qué tan complejo es su borde. En conjunto, estas medidas dan un retrato bastante completo de cómo luce una célula, y los tumores malignos tienden a tener núcleos más grandes, más irregulares y menos simétricos que los benignos.

## *¿Qué se quiere analizar?*
Lo que se busca es responder una pregunta médica muy concreta: ¿se puede predecir si un tumor es maligno o benigno a partir de las características de sus células, sin necesidad de esperar más resultados clínicos.
Para eso se construye un modelo de regresión logística que aprenda de los casos ya diagnosticados y sea capaz de clasificar casos nuevos. También interesa saber qué variables son las más importantes para tomar esa decisión, es decir, cuáles características celulares son las que más distinguen a un tumor maligno de uno benigno.

## *¿Qué variables se tienen que transformar para poder usarse en un modelo de regresión?*
Hay dos variables que necesitan atención especial antes de meterlas al modelo. La primera es diagnosis, que actualmente está guardada como texto con las letras M y B, y los modelos matemáticos no entienden letras, solo números.
La segunda es id, que es simplemente un número de identificación del paciente y no tiene ningún valor predictivo, así que hay que eliminarlo para que no confunda al modelo. Las 30 variables numéricas, en cambio, sí pueden usarse, pero algunas tienen escalas muy diferentes entre sí y otras tienen distribuciones muy asimétricas, lo que también hay que corregir.

## *¿Qué transformaciones se van a usar?*

- Maligno = 1
- Benigno = 0


​​Para diagnosis, la transformación es sencilla: se convierte M en 1 y B en 0, un proceso llamado codificación binaria. La columna id simplemente se elimina del análisis. Para las variables numéricas, se aplica una normalización, que es una forma de estandarizar todas las variables para que tengan la misma escala y ninguna domine al modelo solo por tener números más grandes.

Y para las variables que tienen una distribución muy sesgada hacia la derecha, como la concavidad o la compacidad, se puede aplicar una transformación logarítmica que aplana esa curva y hace que los datos se comporten de forma más equilibrada.

Utilizamos ClassifierMixin, ya que actúa como el mecanismo que vuelve a tu clase compatible con el ecosistema de sklearn. Y BaseEstimator permite que la clase se comporte como un estimador configurable, cuyos hiperparámetros pueden ser leídos y cambiados por una búsqueda automática. ClassifierMixin hace que esa clase sea reconocida específicamente como clasificador. Juntos permiten que BayesSearchCV tome tu clase, la clone, cambie layer1, layer2 y layer3, la entrene y la evalúe correctamente con métricas de clasificación.

## Estandarización y Transformación

```python
num_transformer = StandardScaler()
pt = PowerTransformer(method='yeo-johnson')
num_pipeline = Pipeline(steps=[('scaler', num_transformer), ('power_transformer', pt)])
preprocessor = ColumnTransformer(transformers=[('num', num_pipeline, columnas_numericas)])
```

## **Modelo 1 — Regresión Logística con Regularización**

```python
valores_C = [0.01, 0.1, 1, 10, 100]
kf = KFold(n_splits=10, shuffle=True, random_state=42)
X_sim = np.linspace(0.01, 100, 10000).reshape(-1, 1) # Crea 10,000 números entre 0.01 y 100
```

```python
# REGRESIÓN LOGÍSTICA CON L1

penalty_l1 = 'l1' # Lasso
solver_l1 = 'liblinear' # es el solver compatible con lasso
X_gp_l1 = np.array(valores_C).reshape(-1, 1)

for veces_l1 in range(10):
    resultados_l1 = [] # guardar los AUC de cada C

    for c_l1 in X_gp_l1: # Itera sobre cada valor de C que tenemos
        m_l1 = Pipeline(steps=[
            # estandariza los datos
            ('procesador', preprocessor),
            # entrena una regresión logística con el C actual
             ('modelo', LogisticRegression(C=c_l1[0], penalty=penalty_l1, solver=solver_l1,max_iter=1000, random_state=42))])

        resultados_l1.append(cross_val_score(m_l1, X, y, cv=kf, scoring='roc_auc').mean())
        # Evalúa el modelo con validación cruzada de 10 folds para calcula el AUC promedio y lo guarda en la lista resultados_l1

    gp_l1 = GaussianProcessRegressor(random_state=42) # modelo de Proceso Gaussiano que aprende la relación del valor de C → con el AUC obtenido
    gp_l1.fit(X_gp_l1, np.array(resultados_l1).reshape(-1, 1)) # Entrena el GP con los valores de C y sus AUC correspondientes

    # El GP predice el AUC para los 10,000 valores de C imaginarios
    y_sim_l1, y_std_l1 = gp_l1.predict(X_sim, return_std=True)
    # y_sim_l1 = el AUC predicho para cada valor de C
    # y_std_l1 = la incertidumbre de esa predicción

    # Busca el C con el intervalo superior más alto, le punto más alto
    mejor_C_l1 = X_sim[np.argmax(y_sim_l1 + 2 * y_std_l1)][0]

    print(f'Iteración {veces_l1+1:2d} -> Mejor C con L1: {mejor_C_l1:.4f}')
    X_gp_l1 = np.vstack((X_gp_l1, [[mejor_C_l1]])) # Agrega el nuevo C encontrado a la lista de valores evaluados.

# Guardar el último mejor C encontrado para L1
mejor_C_final_l1 = mejor_C_l1

# AQUI NOS ENFOCAMOS ÚNICAMENTE EL MEJOR C ENCONTRADO
# Entrenar modelo final L1 con el mejor C
modelo_final_l1 = Pipeline(steps=[
    ('procesador', preprocessor),
    ('modelo', LogisticRegression(C=mejor_C_final_l1, penalty='l1', solver='liblinear',max_iter=1000, random_state=42))])

scores_l1 = cross_val_score(modelo_final_l1, X, y, cv=kf, scoring='roc_auc')
auc_medio_l1 = scores_l1.mean()

print(f"Mejor C: {mejor_C_final_l1:.4f}")
print(f"AUC medio (K-Fold): {auc_medio_l1:.4f}")
print(f"Desviación: {scores_l1.std():.4f}")
```

```python
# Gráfica L1
plt.figure(figsize=(10, 6))
plt.plot(X_sim, y_sim_l1, 'b-', label='AUC estimado', linewidth=2)
plt.fill_between(X_sim.ravel(), y_sim_l1.ravel() - 2*y_std_l1.ravel(),y_sim_l1.ravel() + 2*y_std_l1.ravel(), alpha=0.3, color='blue',
                 label='Intervalo de confianza (2σ)') # Esa área representa la incertidumbre de la predicción

plt.scatter(valores_C, resultados_l1[:5], color='red', s=100, zorder=5,label='Puntos evaluados (C iniciales)')

plt.axvline(x=mejor_C_final_l1, color='green', linestyle='--', linewidth=2,label=f'Mejor C = {mejor_C_final_l1:.4f}')
# línea vertical en el mejor C encontrado

plt.xlabel('C (inverso de la regularización)', fontsize=12)
plt.ylabel('AUC (Área bajo la curva ROC)', fontsize=12)
plt.title('Optimización de Regresión Logística L1', fontsize=14, fontweight='bold')
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.xscale('log')
plt.tight_layout()
plt.show()
```

```python
# PARTE 2: REGRESIÓN LOGÍSTICA CON L2

penalty_l2 = 'l2'
solver_l2 = 'lbfgs'
X_gp_l2 = np.array(valores_C).reshape(-1, 1)

for veces_l2 in range(10):
    resultados_l2 = []

    for c_l2 in X_gp_l2:
        m_l2 = Pipeline(steps=[('procesador', preprocessor),
            ('modelo', LogisticRegression(C=c_l2[0], penalty=penalty_l2, solver=solver_l2,max_iter=1000, random_state=42))])

        resultados_l2.append(cross_val_score(m_l2, X, y, cv=kf, scoring='roc_auc').mean())

    gp_l2 = GaussianProcessRegressor(random_state=42)
    gp_l2.fit(X_gp_l2, np.array(resultados_l2).reshape(-1, 1))

    y_sim_l2, y_std_l2 = gp_l2.predict(X_sim, return_std=True)
    mejor_C_l2 = X_sim[np.argmax(y_sim_l2 + 2 * y_std_l2)][0]

    print(f'Iteración {veces_l2+1:2d} -> Mejor C con L2: {mejor_C_l2:.4f}')
    X_gp_l2 = np.vstack((X_gp_l2, [[mejor_C_l2]]))

# Guardar el último mejor C encontrado para L2
mejor_C_final_l2 = mejor_C_l2

# Entrenar modelo final L2 con el mejor C
modelo_final_l2 = Pipeline(steps=[('procesador', preprocessor),
    ('modelo', LogisticRegression(C=mejor_C_final_l2, penalty='l2', solver='lbfgs',max_iter=1000, random_state=42))])

scores_l2 = cross_val_score(modelo_final_l2, X, y, cv=kf, scoring='roc_auc')
auc_medio_l2 = scores_l2.mean()

print(f"Mejor C: {mejor_C_final_l2:.4f}")
print(f"AUC medio (K-Fold): {auc_medio_l2:.4f}")
print(f"Desviación: {scores_l2.std():.4f}")
```

```python
# Gráfica L2
plt.figure(figsize=(10, 6))
plt.plot(X_sim, y_sim_l2, 'b-', label='AUC estimado', linewidth=2)
plt.fill_between(X_sim.ravel(), y_sim_l2.ravel() - 2*y_std_l2.ravel(),
                 y_sim_l2.ravel() + 2*y_std_l2.ravel(), alpha=0.3, color='blue',
                 label='Intervalo de confianza (2σ)')
plt.scatter(valores_C, resultados_l2[:5], color='red', s=100, zorder=5,
           label='Puntos evaluados (C iniciales)')
plt.axvline(x=mejor_C_final_l2, color='green', linestyle='--', linewidth=2,
           label=f'Mejor C = {mejor_C_final_l2:.4f}')
plt.xlabel('C (inverso de la regularización)', fontsize=12)
plt.ylabel('AUC (Área bajo la curva ROC)', fontsize=12)
plt.title('Optimización de Regresión Logística L2', fontsize=14, fontweight='bold')
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.xscale('log')
plt.tight_layout()
plt.show()
```

```python
# Determinar mejor modelo
if auc_medio_l2 > auc_medio_l1:
    print(f"\n MEJOR MODELO: L2 con C = {mejor_C_final_l2:.4f}")
    print(f"   AUC L2: {auc_medio_l2:.4f} | AUC L1: {auc_medio_l1:.4f}")
    print(f"   Mejora: {(auc_medio_l2 - auc_medio_l1)*100:.2f}%")
else:
    print(f"\n MEJOR MODELO: L1 con C = {mejor_C_final_l1:.4f}")
    print(f"   AUC L1: {auc_medio_l1:.4f} | AUC L2: {auc_medio_l2:.4f}")
    print(f"   Mejora: {(auc_medio_l1 - auc_medio_l2)*100:.2f}%")

# Score en test con el mejor modelo
if auc_medio_l2 > auc_medio_l1:
    modelo_final_l2.fit(X_train, y_train)
    score_test = modelo_final_l2.score(X_test, y_test)
    print(f"\nScore en test (L2): {score_test:.4f}")
else:
    modelo_final_l1.fit(X_train, y_train)
    score_test = modelo_final_l1.score(X_test, y_test)
    print(f"\nScore en test (L1): {score_test:.4f}")
```

```python
valores_C_svm = [0.01, 1, 100]
# optimizar gamma
valores_gamma_svm = [0.001, 0.1, 10]
```

```python
X_sim_C = np.linspace(0.01, 100, 10000).reshape(-1, 1)
X_sim_gamma = np.linspace(0.001, 10, 10000).reshape(-1, 1)

X_sim_svm = np.column_stack((X_sim_C, X_sim_gamma))
X_sim_svm
```

```python
# Optimización Bayesiana para la mejor C de SVM con kernel RBF
resultados_svm = []

X_gp_svm = []
for c in valores_C_svm:
    for gamma in valores_gamma_svm:
        model_svm = SVC(kernel='rbf', C=c, gamma=gamma,  random_state=42)
        pipeline_svm = Pipeline(steps=[('procesador', preprocessor), ('modelo', model_svm)])
        cross_val_avg = cross_val_score(pipeline_svm, X, y, cv=kf, scoring='roc_auc').mean()
        X_gp_svm.append([c, gamma])
        resultados_svm.append(cross_val_avg)
```

```python
kernel = 1.0 * RBF(length_scale=1.0); gp_svm = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, random_state=42)

for i in range(30) :

    gp_svm.fit(X_gp_svm, np.array(resultados_svm))

    y_sim_svm, y_std_svm = gp_svm.predict(X_sim_svm, return_std=True)
    y_upper = y_sim_svm + 2 * y_std_svm
    best_i = np.argmax(y_upper)
    mejor_C_svm = X_sim_svm[best_i][0]
    mejor_gamma_svm = X_sim_svm[best_i][1]
    X_gp_svm = np.vstack((X_gp_svm, [mejor_C_svm, mejor_gamma_svm]))

    if i in [0, 29]:
        # Grafica de la función
        # Estilo
        sns.set_theme(style='whitegrid')
        # Crear la figura y el eje 3D
        fig = plt.figure(figsize=(10, 7))
        ax = fig.add_subplot(111, projection='3d')

        # Graficar los puntos evaluados
        surf = ax.scatter(
                X_sim_svm[:, 0],
                X_sim_svm[:, 1],
                y_sim_svm,
                c=y_sim_svm,
                cmap='viridis',
                alpha=0.5)
        # Etiquetas y título
        ax.set_xlabel('C')
        ax.set_ylabel('Gamma')
        ax.set_zlabel('AUC')
        ax.set_title('Grafica inicial' if i == 0 else 'Grafica final')
        ax.legend()
        plt.show()

    model_svm = SVC(kernel='rbf', C=mejor_C_svm, gamma=mejor_gamma_svm, random_state=42)
    pipeline_svm = Pipeline(steps=[('procesador', preprocessor), ('modelo', model_svm)])
    cross_val_avg = cross_val_score(pipeline_svm, X, y, cv=kf, scoring='roc_auc').mean()
    resultados_svm.append(cross_val_avg)

    print(f'Mejor C para SVM RBF: {mejor_C_svm:.4f}, gamma: {mejor_gamma_svm:.4f}, AUC: {cross_val_avg:.4f}')
```

```python
best_score = max(resultados_svm)
best_C_and_gamma = X_gp_svm[np.argmax(resultados_svm)]

print(f'Mejor C: {best_C_and_gamma[0]}, gamma: {best_C_and_gamma[1]}, AUC: {best_score}')
```

## **Modelo 3 — Perceptrón Multicapa (MLP)**

```python
class MLP_Optimization(ClassifierMixin, BaseEstimator):
    '''Se define una clase personalizada compatible con scikit-learn.
       Hereda de ClassifierMixin para que sklearn la reconozca como clasificador,
       y de BaseEstimator para que BayesSearchCV pueda leer y cambiar sus parámetros.'''

    def __init__(self, layer1=10, layer2=10, layer3=10):
        'Define los hiperparámetros del modelo. Cada parámetro representa el número de neuronas de una capa oculta.'
        self.layer1 = layer1
        # Número de neuronas en la primera capa oculta.

        self.layer2 = layer2
        # Número de neuronas en la segunda capa oculta.

        self.layer3 = layer3
        # Número de neuronas en la tercera capa oculta.

    def fit(self, X, y):
        'Entrena el modelo con los datos X y Y.'
        # Cada vez que BayesSearchCV prueba una combinación nueva de layer1, layer2 y layer3,
        # entra aquí para construir y entrenar un MLPClassifier con esos valores.

        self.model = MLPClassifier(
            hidden_layer_sizes=(self.layer1, self.layer2, self.layer3),
            # hidden_layer_sizes indica cuántas neuronas tendrá cada capa oculta.
            # Por ejemplo, si layer1=20, layer2=10 y layer3=5, la arquitectura será (20, 10, 5).

            max_iter=1000,
            # Máximo número de iteraciones de entrenamiento.
            # Se pone alto para darle al modelo suficiente oportunidad de converger.

            random_state=42
            # Fija la semilla aleatoria para que los resultados sean reproducibles.
        )

        self.model.fit(X, y)
        # Entrena el MLPClassifier real con los datos del fold actual.

        # guardar clases para compatibilidad con sklearn
        self.classes_ = self.model.classes_
        # Se copian las clases detectadas por el modelo entrenado.
        # Algunos scorers y utilidades internas de sklearn esperan que un clasificador tenga este atributo.

        return self
        # Por convención en sklearn, fit debe devolver el propio estimador entrenado.



    def predict(self, X):
        # Usa el modelo entrenado para predecir la clase de cada observación de X.
        return self.model.predict(X)

    def predict_proba(self, X):
        # Devuelve las probabilidades predichas de pertenecer a cada clase.
        # Esto es importante para métricas como ROC AUC.
        return self.model.predict_proba(X)

    def score(self, X, y):
        # Devuelve el score por defecto del MLPClassifier.
        # En clasificación, normalmente este score es accuracy.
        return self.model.score(X, y)
```

```python
# Optimización bayesiana
by = BayesSearchCV(
    estimator=MLP_Optimization(),
    # estimator es el modelo base que BayesSearchCV va a entrenar muchas veces.
    # Aquí se parte de la clase personalizada MLP_Optimization.

    search_spaces={
        'layer1': Integer(8, 128),
        # BayesSearchCV podrá probar valores enteros entre 8 y 128
        # para la primera capa oculta.

        'layer2': Integer(4, 64),
        # Valores enteros entre 4 y 64 para la segunda capa oculta.

        'layer3': Integer(2, 16)
        # Valores enteros entre 2 y 16 para la tercera capa oculta.
    },
    n_iter=30,
    # Número de combinaciones que probará la optimización bayesiana.

    cv=kf,

    scoring='roc_auc',
    # Métrica con la que se comparan las distintas arquitecturas.
    # BayesSearchCV elegirá como mejor la combinación con mayor promedio de roc_auc.

    random_state=42
)
```

```python
# Entrenamiento con todos los datos
by.fit(X, y)
# Aquí comienza la búsqueda.
# En cada una de las 30 iteraciones, BayesSearchCV hace:
# 1. propone una combinación de layer1, layer2 y layer3
# 2. crea un MLP-Wrapper con esos valores
# 3. entrena el modelo en los folds de entrenamiento
# 4. lo evalúa en los folds de validación usando roc-auc
# 5. usa esa información para decidir qué combinación probar después
```

```python
print(f'Mejores hiperparámetros:\n 1º capa: {by.best_params_["layer1"]} neuronas,\n 2º capa: {by.best_params_["layer2"]} neuronas,\n 3º capa: {by.best_params_["layer3"]} neuronas')
print(f'Mejor promedio en AUC: {by.best_score_:.5f}')
```

```python
# Tabla comparactiva con la mejor arquitectura encontrada para MLP
tabla_mlp = pd.DataFrame({
    'Mejor capa 1': [f"{by.best_params_['layer1']} neuronas"],
    'Mejor capa 2': [f"{by.best_params_['layer2']} neuronas"],
    'Mejor capa 3': [f"{by.best_params_['layer3']} neuronas"],
    'Mejor AUC (K-Fold)': [f"{by.best_score_:.5f}"]})

print(tabla_mlp.to_string(index=False))
```

```python
final_model_mlp = MLPClassifier(
    hidden_layer_sizes=(by.best_params_['layer1'], by.best_params_['layer2'], by.best_params_['layer3']),
    max_iter=1000,
    random_state=42
)

final_pipeline_mlp = Pipeline(steps=[('procesador', preprocessor), ('modelo', final_model_mlp)])
final_pipeline_mlp.fit(X_train, y_train)

# recall score para evaluar el modelo final
y_pred = final_pipeline_mlp.predict(X_test)
recall_mlp = recall_score(y_test, y_pred)
print(f'Recall del modelo final: {recall_mlp:.4f}')
```
