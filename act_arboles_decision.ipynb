### **Ejercicio**

1. Tomar $advertising.csv$ y hacer una regresion lineal simple

2. Calcular $R^2$ de (1)

3. Usar un árbol de decisión para regresión sobre el mismo dataset

4. Calcular $R^2$ de (3)

5. Repetir (1-4) con trani-test split 50/50

**Escalamiento no necesario**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
```

Carga del dataset

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

Decisión de variables

```python
X = df.drop(['Unnamed: 0', 'sales'], axis=1)
y = df['sales']
```

Regresión lineal simple

```python
lr = LinearRegression()
```

```python
pipeline = Pipeline(
    steps=[('model', lr)]
)
```

```python
pipeline.fit(X, y)
```

$R^2$

```python
pipeline.named_steps['model'].score(X, y)
```

Árbol de decisión

```python
tree_model = DecisionTreeRegressor(random_state=42)
```

```python
pipeline_for_tr = Pipeline(
    steps=[
        ('model', tree_model)
    ]
)
```

```python
pipeline_for_tr.fit(X, y)
```

$R^2$

```python
pipeline_for_tr.named_steps['model'].score(X, y)
```

Separamiento con train-test 50/50

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42) 
```

Entrenamiento con train (regresión lineal)

```python
pipeline.fit(X_train, y_train)
```

$R^2$ train

```python
pipeline.named_steps['model'].score(X_train, y_train)
```

$R^2$ test

```python
pipeline.named_steps['model'].score(X_test, y_test)
```

Entrenamiento con train (árbol de decisión)

```python
pipeline_for_tr.fit(X_train, y_train)
```

$R^2$ train

```python
pipeline_for_tr.named_steps['model'].score(X_train, y_train)
```

$R^2$ test

```python
pipeline_for_tr.named_steps['model'].score(X_test, y_test)
```
